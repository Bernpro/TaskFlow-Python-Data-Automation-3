```python
import os
import time
import pandas as pd
import requests
from bs4 import BeautifulSoup
from openpyxl import load_workbook
from openpyxl.styles import Alignment, Font, PatternFill

# Para el Bot de Tareas usaremos Selenium (controlador del navegador)
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager

# =====================================================================
# MÓDULO 1: WEB SCRAPING (Extraer datos de un proveedor simulado)
# =====================================================================
def ejecutar_scraping():
    print("🕵️‍♂️ [PASO 1]: Extrayendo datos de competidores...")

    # Datos simulados de productos que el scraping recolectó de la web
    datos_web = [
        {"Producto": "Laptop Premium X", "Precio_Base": 1200},
        {"Producto": "Monitor Gamer 4K", "Precio_Base": 450},
        {"Producto": "Teclado Mecanico RGB", "Precio_Base": 90}
    ]

    df = pd.DataFrame(datos_web)
    archivo_crudo = "datos_crudos.xlsx"
    df.to_excel(archivo_crudo, index=False)
    print("✅ Scraping finalizado. Datos guardados.")
    return archivo_crudo

# =====================================================================
# MÓDULO 2: AUTOMATIZACIÓN DE REPORTES (Convertir Excel aburrido en ejecutivo)
# =====================================================================
def generar_reporte(archivo_entrada):
    print("\n📊 [PASO 2]: Procesando cálculos y dando diseño al Excel...")

    # Leer el excel crudo del scraping y calcular impuestos automáticos
    df = pd.read_excel(archivo_entrada)
    df["IVA (16%)"] = df["Precio_Base"] * 0.16
    df["Precio_Final"] = df["Precio_Base"] + df["IVA (16%)"]

    archivo_reporte = "Reporte_Final_Ejecutivo.xlsx"
    df.to_excel(archivo_reporte, index=False, sheet_name="Precios_Monitoreados")

    # Diseño estético corporativo con OpenPyXL
    wb = load_workbook(archivo_reporte)
    ws = wb["Precios_Monitoreados"]

    azul_encabezado = PatternFill(start_color="1F4E78", end_color="1F4E78", fill_type="solid")
    texto_blanco = Font(name="Arial", size=11, bold=True, color="FFFFFF")

    # Aplicar estilos a la primera fila (Encabezados)
    for col in range(1, ws.max_column + 1):
        celda = ws.cell(row=1, column=col)
        celda.fill = azul_encabezado
        celda.font = texto_blanco
        celda.alignment = Alignment(horizontal="center")

    # Dar formato de dinero ($) a las columnas numéricas
    for fila in range(2, ws.max_row + 1):
        ws.cell(row=fila, column=2).number_format = "$#,##0.00"
        ws.cell(row=fila, column=3).number_format = "$#,##0.00"
        ws.cell(row=fila, column=4).number_format = "$#,##0.00"

    wb.save(archivo_reporte)
    print("✅ Reporte ejecutivo formateado correctamente.")
    return archivo_reporte

# =====================================================================
# MÓDULO 3: BOT DE TAREAS (Escribir automáticamente en un sistema web)
# =====================================================================
def ejecutar_bot_de_tareas(archivo_excel):
    print("\n🤖 [PASO 3]: Iniciando Bot de Tareas para subir información...")

    # El bot lee el reporte inteligente que creamos en el Paso 2
    df = pd.read_excel(archivo_excel)

    # Configurar y abrir el navegador Chrome real de tu computadora
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)

    try:
        # Entramos a una página web real de pruebas para inicio de sesión corporativo
        driver.get("https://herokuapp.com")
        driver.maximize_window()
        time.sleep(2) # Espera visual de cortesía

        # --- AQUÍ OPERA EL BOT DE TAREAS ---
        print("-> El bot está escribiendo las credenciales de acceso automáticamente...")
        campo_usuario = driver.find_element(By.ID, "username")
        campo_password = driver.find_element(By.ID, "password")
        boton_entrar = driver.find_element(By.CSS_SELECTOR, "button[type='submit']")

        # El bot toma el nombre del primer producto del Excel y lo usa como usuario simulado
        producto_excel = df.iloc["Producto"]

        campo_usuario.send_keys(producto_excel) # El bot tipea en la pantalla solo
        time.sleep(1)
        campo_password.send_keys("SuperSecretPassword!") # Tipea la contraseña corporativa
        time.sleep(1)

        print("-> El bot hace clic en el botón de envío por ti...")
        boton_entrar.click() # El mouse hace clic automáticamente
        time.sleep(4) # Esperar a ver el resultado en pantalla

        print("✅ Formulario rellenado y enviado con éxito por el Bot.")

    except Exception as e:
        print(f"❌ Ocurrió un error en el Bot de Tareas: {e}")
    finally:
        # Cerrar el navegador controlado por el robot al terminar
        driver.quit()

# =====================================================================
# ORQUESTADOR (Une las 3 piezas en cadena)
# =====================================================================
if __name__ == "__main__":
    print("🚀 ARRANCANDO SOFTWARE INTEGRADO DE AUTOMATIZACIÓN 🚀\n")

    archivo_scraping = ejecutar_scraping()
    archivo_final_excel = generar_reporte(archivo_scraping)
    ejecutar_bot_de_tareas(archivo_final_excel)

    print("\n🎉 ¡Las 3 automatizaciones se ejecutaron consecutivamente con éxito!")
```
